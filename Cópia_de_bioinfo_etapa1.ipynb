```python
# Contar variantes por impacto
VCF_ANN = "bioinfotest/output/10-7-BRCA.annotated.vcf"

print("=== CONTAGEM DE VARIANTES POR IMPACTO ===\n")

# HIGH
!echo "Variantes HIGH:"
!grep -v "^#" $VCF_ANN | grep -o "HIGH" | wc -l

# MODERATE
!echo "Variantes MODERATE:"
!grep -v "^#" $VCF_ANN | grep -o "MODERATE" | wc -l

# LOW
!echo "Variantes LOW:"
!grep -v "^#" $VCF_ANN | grep -o "LOW" | wc -l

print("\n=== DETALHES DAS VARIANTES HIGH ===\n")
!echo "=== Variantes HIGH ==="
!grep "HIGH" $VCF_ANN | grep -v "^#" | cut -f1,2,4,5
# Mostrar detalhes das variantes HIGH
!echo ""
!echo "=== Detalhes das variantes HIGH ==="
!grep -v "^#" $VCF_ANN | grep "HIGH" | awk '{print "Cromossomo: "$1"\nPosição: "$2"\nRef: "$4"\nAlt: "$5"\n---"}'
```

```python
# 1. Verificar qual arquivo BED você está usando
!echo "=== Conteúdo do alvos.bed (ERRADO) ==="
!cat bioinfotest/data/alvos.bed

# 2. Criar o arquivo BED CORRETO com BRCA1 e BRCA2
with open('bioinfotest/data/BRCA.bed', 'w') as f:
    f.write("chr17\t41196312\t41277500\n")  # BRCA1
    f.write("chr13\t32889617\t32973809\n")  # BRCA2

!echo -e "\n=== Conteúdo do BRCA.bed (CORRETO) ==="
!cat bioinfotest/data/BRCA.bed

# 3. Verificar se há cobertura nas regiões BRCA
BAM = "bioinfotest/output/10-7-BRCA.sorted.bam"
!echo -e "\n=== Cobertura em BRCA1 ==="
!samtools view -c $BAM chr17:41196312-41277500

!echo -e "\n=== Cobertura em BRCA2 ==="
!samtools view -c $BAM chr13:32889617-32973809

# 4. RE-EXECUTAR FreeBayes com o arquivo CORRETO
REF = "bioinfotest/data/hg19.fasta"
BED = "bioinfotest/data/BRCA.bed"  # ARQUIVO CORRETO!
VCF_OUT = "bioinfotest/output/10-7-BRCA.raw.vcf"

!freebayes -f $REF --targets $BED $BAM > $VCF_OUT

# 5. Verificar quantas variantes foram encontradas
!echo -e "\n=== Variantes encontradas no VCF raw ==="
!grep -v "^#" $VCF_OUT | wc -l
!grep -v "^#" $VCF_OUT | head -5

# 6. Re-anotar com SnpEff
VCF_ANN = "bioinfotest/output/10-7-BRCA.annotated.vcf"
!java -Xmx4g -jar snpEff/snpEff.jar hg19 $VCF_OUT > $VCF_ANN

# 7. Verificar variantes anotadas
!echo -e "\n=== Variantes anotadas ==="
!grep -v "^#" $VCF_ANN | wc -l
```
